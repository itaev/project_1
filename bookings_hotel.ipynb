```python
#импортируем необходимые библиотеки
import pandas as pd
```

```python
#читаем bookings.csv, сохраняем в переменную bookings
bookings = pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-m-itaev-23/mini_project_1/bookings.csv', sep=';')
```

# Задача 2

2. Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.

```python
#задаем функцию для приведения наименований к нижнему регистру и замене пробелов на нижние подчеркивания
def replace_new_columns(columns):
    new_name = columns.lower().replace(' ', '_')
    return new_name
```

```python
#заменяем наименования в нашей таблице
bookings = bookings.rename(columns = replace_new_columns)
bookings.head()
```

# Задача 3

3. Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
# отфильтровали по бронированиям, которые не были отменены
# сгруппировали по странам
# посчитали количество успешных бронирований
# отсортировали по убыванию
# вывели ТОП 5 стран

top_canceled = bookings.query('is_canceled == 0') \
    .groupby('country', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(5)
top_canceled
```

# Задача 4

4. На сколько ночей в среднем бронируют отели разных типов?

```python
# сгруппировали по отелям
# посчитали среднее число общих заброннированных ночей
# округлили значения до 2-х знаков

top_hotel = bookings.groupby('hotel', as_index=False) \
    .agg({'stays_total_nights': 'mean'}) \
    .round(2)
top_hotel
```


# Задача 5

# Задача 6

6. Проанализируйте даты запланированного прибытия. – На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?– Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов

```python
#отсортировали таблицу по 2016 году
#сгруппировали по году и месяцу
#посчитали количество по месяцам 2016 года
#отсортирвали по убыванию
top_arrival_date_2016 = bookings.query('arrival_date_year == 2016') \
                            .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
                            .agg({'hotel': 'count'}) \
                            .sort_values('hotel',  ascending=False)
top_arrival_date_2016.head(5)
```

```python
#отсортировали планируемую дату прилета за 2017г
#сгруппировали по году и месяцу
#посчитали количество броней и отсортировали по убыванию
#сравниваем с данным выше за 2016г.
top_arrival_date_2017 = bookings.query('arrival_date_year == 2017') \
                            .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
                            .agg({'hotel': 'count'}) \
                            .sort_values('hotel',  ascending=False)
top_arrival_date_2017.head(5)
```

```python
#вывели две таблицы, в которых данные сгруппированы по годам и месяцам по двум отелям City Hotel и Resort Hotel
#выведено количество бронирований по указанной группироваке
hotel_city = bookings.query('hotel == City Hotel' and 'is_canceled == 1') \
                    .groupby(['arrival_date_year', 'arrival_date_month']) \
                    .agg({'is_canceled': 'count'}) \
                    .sort_values(['arrival_date_year', 'is_canceled'], ascending=[True, False])
hotel_resort = bookings.query('hotel == Resort Hotel' and 'is_canceled == 1') \
                    .groupby(['arrival_date_year', 'arrival_date_month']) \
                    .agg({'is_canceled': 'count'}) \
                    .sort_values(['arrival_date_year', 'is_canceled'], ascending=[True, False])
```

# Задача 8

8. Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
#вывели таблицу только с интересующими нас столбцами
total_kids_df = bookings[['hotel', 'is_canceled', 'adults', 'children', 'babies']]
total_kids_df
```

```python
# добавили новый столбец, в котором суммированы значения по столбцами children и babies
total_kids_df['total_kids'] = total_kids_df['children'] + total_kids_df['babies']
```

```python
# сгруппировали по отелям и посчитали среднее количество детей, которое указывали при бронировании
total_kids_df.groupby('hotel', as_index=False) \
            .agg({'total_kids': 'mean'}) \
            .round(2)
```

# Задача 9

9. Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди какой группы показатель выше.

```python
#добавили новый столбец в исходную таблицу с общим количеством детей
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
#добавили новый столбец, который отражает наличие детей при бронировании номера
bookings['has_kids'] = bookings['total_kids'] > 0
```

```python
#проверяем результат
bookings.head()
```

```python
#посчитали, сколько бронирований было отменено и прошло успешло в бронях, где были деди
has_kind_canceled = bookings.query('has_kids == True').is_canceled.value_counts(normalize=True) * 100
```

```python
#округлили значения до 2-х знаков
has_kind_canceled.round(2)
```

```python
#посчитали, сколько бронирований было отменено и прошло успешло в бронях, где детей не было
not_has_kind_canceled = bookings.query('has_kids == False').is_canceled.value_counts(normalize=True) * 100
```

```python
#округлили значения до 2-х знаков
not_has_kind_canceled.round(2)
```
